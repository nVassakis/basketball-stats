```python
import pandas as pd
import json

# 1. Load the conversion JSON
with open('teams.json', 'r', encoding='utf-8') as f:
    teams_data = json.load(f)

print(f"Loaded {len(teams_data)} teams from JSON.")

# 2. Load CSV
# I removed '.iloc[:, :10]' because the 'Team' column is likely at the end of your file.
# We need to read the whole file to modify the 'Team' column and keep the other data.
df = pd.read_csv('/Users/nikos/basketball-stats/data/processed/teams_stats_master.csv')

# 3. Search for 'Team' and make the conversion
if 'Team' in df.columns:
    def convert_team(val):
        val_str = str(val)
        # Check every slug in the dictionary against the cell value
        for slug, name in teams_data.items():
            if slug in val_str:
                return name
        return val # Keep original if no match found

    df['Team'] = df['Team'].apply(convert_team)
    print("✅ Conversion applied successfully.")
else:
    print("⚠️ Warning: 'Team' column not found in the CSV. No changes made.")

# 4. Output the new CSV
df.to_csv('updated_teams.csv', index=False)
print("Done! Saved as 'updated_teams.csv'")
```
